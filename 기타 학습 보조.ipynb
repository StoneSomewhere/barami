```python
#파일 목록 csv로 저장

import os
import pandas as pd
import numpy as np

path_dir = './Train_1h/'

folder_list = os.listdir(path_dir)

for folder in folder_list:
    if ".ini" in folder: continue
    DF = pd.DataFrame()
    print(folder)
    
    file_list = os.listdir(path_dir+folder)
    file_list = [x for x in file_list if "1h" in x]
    print(file_list)
    DF[folder] = file_list
    print(DF)
    DF.to_csv("%s_list.csv"%folder, mode='w')
```
